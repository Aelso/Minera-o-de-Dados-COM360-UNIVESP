**Comentário da resposta:**

A correlação é positiva quando os valores relacionados são diretamente proporcionais, e é negativa quando esses valores são inversamente proporcionais. A ausência de correlação ocorre quando os dados estão dispersos, sem um padrão de relacionamento entre eles. A correlação não linear indica algum padrão de correlação parcial.
